```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xp=np.array([1,10]) #line plotting
yp=np.array([3,10])
plt.plot(xp,yp)
plt.show()
```

```python
plt.plot(xp,yp,'o') #without line
plt.show()
```

```python
xp=np.array([1,2,6,9]) #multiple points
yp=np.array([1,4,7,3])
plt.plot(xp,yp)
plt.show()
```

```python
 #x-points by default
yp=np.array([3,8,1,10,5,7])
plt.plot(yp)
plt.show()
```

```python
xp=np.array([1,2,6,9]) #marker-point
yp=np.array([1,4,7,3])
plt.plot(xp,yp, marker='+')
plt.show()
```

```python
 #apply various color of marker
yp=np.array([3,8,1,10,5,7])
plt.plot(yp,'o:r')
plt.show()
```

```python
 #edge color and marker color
yp=np.array([3,8,1,10,5,7])
plt.plot(yp,marker='*',mec='g',mfc='r')
plt.show()
```

```python
#apply various line type and color
yp=np.array([3,8,1,10,5,7])
plt.plot(yp,'o:') #ls='--'
plt.show()
```

```python
#apply various linewidth
yp=np.array([3,8,1,10,5,7])
plt.plot(yp,linewidth='10')
plt.show()
```

```python
#apply various multiple lines at a time 
x1=np.array([1,2,3,10])
x2=np.array([0,7,8,9])
y1=np.array([3,0,8,5])
y2=np.array([5,4,2,1])
plt.plot(x1,y1,x2,y2)
plt.show()
```

```python
#Labels and title
xp=np.array([1,2,4,8])
yp=np.array([3,8,6,9])
plt.plot(xp,yp) 
plt.xlabel('india gdp')
plt.ylabel('growth')
plt.show()
```

```python
xp=np.array([1,2,3,10])
yp=np.array([3,8,1,10])  #grid
plt.title('john')
plt.xlabel('rate')
plt.ylabel('profit')
plt.plot(xp,yp)
plt.grid(color='green')
plt.show()
```

```python
xp=np.array([1,2,3,10])
yp=np.array([3,8,1,10]) #subplot
plt.subplot(2,2,1)
plt.plot(xp,yp)


x=np.random.normal(170,10,250) #histor
plt.subplot(2,2,2)
plt.hist(x)

xp=np.array(['a','b','c'])
yp=np.array([4,1,7]) #bar
plt.subplot(2,2,3)
plt.bar(xp,yp)

y=np.array([80,15,5])
mylabels=['ycp','tdp','jsp']
plt.subplot(2,2,4)
plt.title('survey')

plt.pie(y,labels=mylabels,autopct='%1.1f%%',explode=[0,0,1])
plt.show()
```

```python
xp=np.array([0,1,2,3])
yp=np.array([2,3,1,9])#subplot
plt.subplot(1,2,1)
plt.plot(xp,yp)

xp=np.array([1,2,3,10])
yp=np.array([4,7,2,8])#subplot
plt.subplot(1,2,2)
plt.plot(xp,yp)
plt.show()
```

```python
xp=np.array([0,1,2,3])
yp=np.array([2,4,6,8])#scatter
plt.scatter(xp,yp)
plt.show()
```

```python
xp=np.array([1,2,3,10])
yp=np.array([3,8,1,10])  #scatter_2
plt.scatter(xp,yp,color='red')

xp=np.array([1,2,3,10])
yp=np.array([4,7,2,8])
plt.scatter(xp,yp,color='green')

plt.show()
```

```python
x=np.array(['Maths','english','social']) #bar graph
y=np.array([5,8,10])
plt.bar(x,y,color='green',width=0.5)
plt.show()
```
